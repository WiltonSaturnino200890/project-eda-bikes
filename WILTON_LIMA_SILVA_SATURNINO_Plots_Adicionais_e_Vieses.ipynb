Usando scatter plot polar: https://matplotlib.org/stable/gallery/pie_and_polar_charts/polar_scatter.html

```python
import numpy as np
import matplotlib.pyplot as plt

# Criar num de pontos
# para simular os crculos
N = 150
r = 2 * np. random.rand(N)
theta = 2 * np.pi * np.random.rand(N)

# Calcular a rea de cada crculo
area = 200 * r**2
colors = theta

fig = plt.figure()

# Grfico em coordenadas polares
ax = fig.add_subplot(projection = 'polar')
c = ax.scatter(theta, r, c=colors, s=area, cmap= 'nipy_spectral', alpha= 0.75)
ax.set_rlabel_position(225)
plt.yticks(size=6)
# plt.yticks([0.50, 1.00, 1.50, 2.0], ["0.50", "1.00", "1.50", "2.0"], size=7)
# plt.yticks([0.25, 0.50, 0.75, 1.00, 1.25, 1.50, 1.75, 2.0], ["0.25", "0.50", "0.75", "1.00", "1.25", "1.50", "1.75", "2.0"], color="grey", size=7)
# plt.ylim(0,2)
plt.show()
```

```python
fig = plt.figure()
ax= fig.add_subplot(projection='polar')

# Aqui, h a rotao do ponto inicial do raio
ax.set_rorigin(-2.5)
c = ax.scatter(theta, r, c=colors, s=area, cmap= 'nipy_spectral', alpha = 0.75)
# Aqui, h a rotao grfico
ax.set_theta_zero_location('W', offset=10)

ax.set_rlabel_position(225)
plt.yticks(size=6)
plt.show()
```

Plot polar de barras: https://matplotlib.org/stable/gallery/pie_and_polar_charts/polar_bar.html

```python
import numpy as np
import matplotlib.pyplot as plt

# Nmero de setores no grfico
N = 20
theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)

# Raio de cada setor do grfico
radii = 10 * np.random.rand(N)

# ngulo de abertura de cada setor (largura)
width = np.pi / 4 * np.random.rand(N)
colors = plt.cm.Spectral(radii / 10.)

ax = plt.subplot(projection= 'polar')
ax.bar(theta, radii, width, bottom=2, color=colors, alpha=0.5)
plt.show()
```

Plot tipo radar

```python
from math import pi
import matplotlib.pyplot as plt
```

```python
def RadarChart (player, data, player2, data2):
  Attributes = ["Geral", "Potencial", "Ritmo", "Chute", "Passe", "Drible", "Defesa", "Fsico"]
  
  angles = [n / 8 * 2 * pi for n in range (8)]
  angles += angles[:1]
  
  angles2 = [n / 8 * 2 * pi for n in range(8)]
  angles2 += angles2[:1]

  plt. figure (dpi=125)
  ax = plt.subplot(111, polar=True)
  ax = plt.subplot(111, polar=True)
  plt.xticks(angles [:-1], Attributes)
  
  ax.plot(angles, data)
  ax.fill(angles, data, 'teal', alpha=0.1)
  
  ax.plot(angles2, data2)
  ax.fill(angles2, data2, 'red', alpha=0.1)
  plt.figtext(0.1, 0.98, player, color="steelblue")
  plt.figtext(0.1, 0.94, "vs")
  plt.figtext(0.1, 0.90, player2, color="darkorange")
  plt.show()
```

```python
RadarChart("Messi", [94.0, 94.0, 87.0, 92.0, 92.0, 96.0, 39.0, 66.0, 94.0],
           "Ronaldo", [93.0, 93.0, 90.0, 93.0, 82.0, 89.0, 35.0, 78.0, 93.0])
```
